```python
import requests
import json
import pandas as pd
```

```python
def get_stock_data(id, scale, data_len):
    # 新浪财经接口
    url = 'http://quotes.sina.cn/cn/api/json_v2.php/CN_MarketDataService.getKLineData?symbol={0}&scale={1}&datalen={2}'
    response = requests.get(url.format(id, scale, data_len))
    bar_list = []
    resp_json = response.json()
    resp_json.reverse() # 倒序
    
    for dc in resp_json:
        bar = {}
        bar['date'] = dc['day']
        bar['open'] = float(dc['open'])
        bar['high'] = float(dc['high'])
        bar['low'] = float(dc['low'])
        bar['close'] = float(dc['close'])
        bar['volume'] = float(dc['volume'])
        bar['amount'] = float(dc['amount'])
        bar_list.append(bar)
    return pd.DataFrame(bar_list)
```

```python
get_stock_data('sh000001', 5, 49)
```
